# Homework 1 - Analytical solution of heat equation

The surface heat flow density at the area of interest is between $70$ and $92\;\mathrm{mWm^{-2}}$. The average temperature of the surface is $T_0 = 10\mathrm{°C}$ 

The crustal thickness is $h=34\;\mathrm{km}$, the crustal conductivity is $\lambda = 2.5\;\mathrm{Wm^{-1}°C^{-1}}$ and the crustal radioactive heat production is $H = 1.6\;\;\mathrm{μWm^{-3}}$.

Using analytical solution of steady-state heat equation

 * calculate the temperature at MOHO (for minimum and maximum surface heat flow density).
 * calculate the mantle heat flow at MOHO (for both temperatures calculated at 1).
 * plot geotherms for both surface heat flow densities.

```python
T_0 = 10
cc = 2.5
H = 1.6e-6
crust_thickness_km = 34
crust_thickness_m = crust_thickness_km * 1000


q_min = -70e-3 
q_max = -92e-3  


depth_range = np.linspace(0, crust_thickness_m, 1000)


T_min = (-H / (2 * cc) * depth_range**2 - q_min / cc * depth_range + T_0)
T_max = (-H / (2 * cc) * depth_range**2 - q_max / cc * depth_range + T_0)

q_Moho_min = H*crust_thickness_m + q_min
q_Moho_max = H*crust_thickness_m + q_max


plt.figure(figsize=(6, 4))
plt.plot(T_min, depth_range, label=f"q_min = {q_min*1e3} mW/m^2")
plt.plot(T_max, depth_range, label=f"q_max = {q_max*1e3} mW/m^2")
plt.xlabel('Temperature (°C)')
plt.ylabel('Depth (m)')
plt.title('Geotherms')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()


print("Temperature at MOHO for minimum surface heat flow density: {:.2f} °C".format(T_min[-1]))
print("Temperature at MOHO for maximum surface heat flow density: {:.2f} °C".format(T_max[-1]))
print("Heat flow at MOHO for minimum surface heat flow density: {:.3f} W/m^2".format(q_Moho_min))
print("Heat flow at MOHO for maximum surface heat flow density: {:.3f} W/m^2".format(q_Moho_max))
```
